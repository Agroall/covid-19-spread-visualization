```python
import pandas as pd
```

```python
data = pd.read_csv('cleaned_data')
data.set_index('Country/Region', inplace=True)
data.head()
```

```python
test = data.loc[:, :'May 2020']\
                    .max(axis=1)\
                    .to_frame(name='max_cases')
max_val = test['max_cases'].max()
min_val = test['max_cases'].min()

test['scaled_confirmed_cases'] = 1 + ((test['max_cases'] - min_val) * 9) /(max_val-min_val)
test
```

```python
def month_selector(df, month):
    return df.loc[:, :month]\
                    .max(axis=1)\
                    .to_frame(name='max_cases')
    
    
```

```python
month_selector(data, "May 2020")
```
